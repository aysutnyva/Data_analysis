```python
import numpy as np
import pandas as pd
```

```python
waze = pd.read_csv("/content/waze_dataset.csv.zip")
```

```python
waze.head(10)
```

```python
waze.info()
```

```python
null_waze = waze[waze['label'].isnull()]
waze.describe()
```

```python
not_null_waze = waze[~waze['label'].isnull()]
not_null_waze.describe()
```

```python
null_waze['device'].value_counts()
```

```python
null_waze['device'].value_counts(normalize=True)
```

```python
waze['device'].value_counts(normalize=True)
```

```python
print(waze['label'].value_counts())
print()
print(waze['label'].value_counts(normalize=True))
```

```python
waze.groupby('label').median(numeric_only=True)
```

```python
waze['km_per_drive'] = waze['driven_km_drives'] / waze['drives']
median_km_per_drive = waze.groupby('label').median(numeric_only=True)[['km_per_drive']]
median_km_per_drive
```

```python
waze['km_per_driving_day'] = waze['driven_km_drives'] / waze['driving_days']
median_km_per_driving_day = waze.groupby('label').median(numeric_only=True)[['km_per_driving_day']]
median_km_per_driving_day
```

```python
waze['drives_per_driving_day'] = waze['drives'] / waze['driving_days']
median_drives_per_driving_day = waze.groupby('label').median(numeric_only=True)[['drives_per_driving_day']]
median_drives_per_driving_day
```
